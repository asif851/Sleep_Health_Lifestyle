```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
df.isnull().sum()  
df = df.dropna()  
```

```python
plt.hist(df['Sleep Duration'], bins=20)
plt.xlabel('Sleep Duration')
plt.ylabel('Frequency')
plt.title('Sleep Duration Distrinution')
plt.show()
```

```python
plt.scatter(df['Sleep Duration'], df['Heart Rate'])
plt.xlabel('Sleep Duration')
plt.ylabel('Heart Rate')
plt.title('Sleep Duration V/S Heart Rate')
plt.show()
```

```python
occupation_counts = df['Occupation'].value_counts()
plt.bar(occupation_counts.index, occupation_counts.values)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Sleep Disorders by Occupation')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Age Group'] = pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 60, 100], labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61+'])
average_sleep_duration_by_age = df.groupby('Age Group')['Sleep Duration'].mean()

plt.bar(average_sleep_duration_by_age.index, average_sleep_duration_by_age.values)
plt.xlabel('Age Group')
plt.ylabel('Average Sleep Duration')
plt.title('Average Sleep Duration by Age Group')
plt.show()
```

```python
bmi_values = df['BMI Category']

bmi_counts = bmi_values.value_counts()

plt.figure(figsize=(10, 6))
bmi_counts.plot(kind='bar')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.title('BMI Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.countplot(x='Sleep Disorder', hue='Gender', data=df)
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
plt.title('Sleep Disorder Counting by Gender')
plt.legend(title='Gender')
plt.show()
```

```python
grouped_data = df.groupby(['Gender', 'Blood Pressure']).size().unstack()

plt.figure(figsize=(8, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Blood Pressure')
plt.title('Blood Pressure by Gender')
plt.xticks(rotation=0)
plt.legend(title='Blood Pressure')
plt.show()
```
